# Analyse d'Algorythme

## Bruteforce

Dans l'algorythme bruteforce, on va parcourir toutes les combinaisons possibles. 
On à recourt au notion de dénombrement mathématique, plus précisement la combianaison d'un ensemble d'éléments, soit une disposition non ordonnée d'un certain nombre d'éléments de cet ensemble.

### Formule mathématique: 
Le nombre de combinaisons de **k** éléments parmi **n** éléments se note $\binom{n}{k}$  et s'écrit sous la forme de fraction de produit factoriel :
<img src="https://latex.codecogs.com/gif.latex?\frac{n!}{k!(n-k)!}" title="\frac{n!}{k!(n-k)!}" style="vertical-align: middle;" />

### Représentation graphique du force brut pour 20 éléments :

```python
import math
import matplotlib.pyplot as plt
import numpy as np

NUM_ACTION = 20

results = []
list_sum = []
sum = 0
for k in range(NUM_ACTION):
    result = math.factorial(NUM_ACTION)/(math.factorial(k)*(math.factorial(NUM_ACTION-k)))
    
    sum+= result
    list_sum.append(sum)
    results.append(result)
    

plt.plot(results)
plt.title("Courbe en cloche, représente le nombre de possibilités de combinaisons par rapport aux nombres d'éléments par combinaison")
#plt.plot(list_sum)
```

La courbe en cloche ci-dessus, représente le nombre de possibilités de combinaisons par rapport aux nombres d'éléments par combinaison 

```python
plt.plot(list_sum)

print(list_sum[-1])
```

La courbe ci-dessus représente la somme de toutes les combinaisons en fonctions du nombre d'éléments par combinaison

## Optimized

L'algorithme optimzed.py à une complexité linéaire = O(n.log(n))

```python
plt.plot([x*math.log(x) for x in range(1,NUM_ACTION)], label='optimized complexity')
plt.legend()
```
